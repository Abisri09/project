```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models
import os
from tqdm import tqdm
```

```python
combined_dataset_path = "/content/combined_rice_dataset"

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
```

```python
train_dataset = datasets.ImageFolder(os.path.join(combined_dataset_path), transform=transform)

# Create DataLoader
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4)
```

```python
model = models.efficientnet_b0(pretrained=True)

# Modify the classifier layer to match the number of rice categories
num_ftrs = model.classifier[1].in_features
num_classes = len(train_dataset.classes)  # Number of rice categories
model.classifier[1] = nn.Linear(num_ftrs, num_classes)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
num_epochs = 5
for epoch in range(num_epochs):
    model.train()  # Set the model to training mode
    running_loss = 0.0
    correct_predictions = 0
    total_predictions = 0

    for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}", unit="batch"):
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()  # Zero the parameter gradients

        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        # Calculate running loss and accuracy
        running_loss += loss.item() * inputs.size(0)
        _, predicted = torch.max(outputs, 1)
        correct_predictions += (predicted == labels).sum().item()
        total_predictions += labels.size(0)

    # Calculate average loss and accuracyfor the epoch
    epoch_loss = running_loss / len(train_dataset)
    epoch_acc = 100 * correct_predictions / total_predictions
    print(f"Epoch {epoch+1}/{num_epochs} - Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.2f}%")

# Save the trained model
torch.save(model.state_dict(), "efficientnet_b0_rice_model.pth")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def evaluate_model(model, dataloader, device):
    model.eval()  # Set the model to evaluation mode
    all_labels = []
    all_preds = []

    with torch.no_grad():  # No need to track gradients during evaluation
        for inputs, labels in tqdm(dataloader, desc="Evaluating", unit="batch"):
            inputs, labels = inputs.to(device), labels.to(device)

            # Forward pass
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)

            # Store labels and predictions for metrics
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(predicted.cpu().numpy())

    # Calculate accuracy
    accuracy = sum(np.array(all_preds) == np.array(all_labels)) / len(all_labels) * 100

    # Calculate precision, recall, and F1 score
    precision = precision_score(all_labels, all_preds, average='weighted')
    recall = recall_score(all_labels, all_preds, average='weighted')
    f1 = f1_score(all_labels, all_preds, average='weighted')

    # Confusion matrix
    cm = confusion_matrix(all_labels, all_preds)

    return accuracy, precision, recall, f1, cm
```

```python
validation_dataset = datasets.ImageFolder(os.path.join(combined_dataset_path), transform=transform)
validation_loader = DataLoader(validation_dataset, batch_size=32, shuffle=False, num_workers=4)
```

```python
accuracy, precision, recall, f1, cm = evaluate_model(model, validation_loader, device)

# Print the metrics
print(f"Validation Accuracy: {accuracy:.2f}%")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=validation_dataset.classes, yticklabels=validation_dataset.classes)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
import torch
from torchvision import transforms
from PIL import Image
```

```python
def process_image(image_path, transform):
    image = Image.open(image_path)  # Open the image
    image = transform(image)  # Apply the same transformations as in training
    image = image.unsqueeze(0)  # Add batch dimension
    return image
```

```python
model = models.efficientnet_b0(pretrained=False)  # Don't load pretrained weights
model.classifier[1] = torch.nn.Linear(model.classifier[1].in_features, len(validation_dataset.classes))  # Adjust the classifier to match the number of classes
model.load_state_dict(torch.load("efficientnet_b0_rice_model.pth"))  # Load the trained weights
model = model.to(device)  # Move to GPU if available
model.eval()
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
```

```python
image_path1 = '/content/Ipsala (100).jpg'
image_path2 = '/content/mixed.jpg'
```

```python
image1 = process_image(image_path1, transform)
image2 = process_image(image_path2, transform)
```

```python
image1 = image1.to(device)
image2 = image2.to(device)
```

```python
with torch.no_grad():  # No need to track gradients during inference
    outputs1 = model(image1)
    outputs2 = model(image2)

    _, predicted1 = torch.max(outputs1, 1)
    _, predicted2 = torch.max(outputs2, 1)
```

```python
predicted_class1 = validation_dataset.classes[predicted1.item()]
predicted_class2 = validation_dataset.classes[predicted2.item()]
```
